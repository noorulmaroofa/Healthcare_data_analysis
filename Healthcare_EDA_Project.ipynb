# Healthcare Diabetes EDA Project
Complete Data Cleaning, EDA, and Export for Power BI

```
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load Dataset
df = pd.read_csv('Healthcare-Diabetes.csv')
df.head()
```

```
# Basic Info
df.info()
df.describe()
```

```
# Replace invalid zeros with NaN
cols = ["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
for col in cols:
    df[col] = df[col].replace(0, np.nan)

# Fill missing values
df.fillna(df.median(numeric_only=True), inplace=True)

df.isnull().sum()
```

```
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```
# Univariate Analysis
df.hist(figsize=(10,8))
plt.show()

sns.boxplot(data=df)
plt.show()
```

```
# Bivariate Analysis
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df)
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```
# Multivariate Analysis
sns.pairplot(df, hue='Outcome')
plt.show()
```

```
# Grouping & Aggregation
df.groupby('Outcome').mean()
```

```
# Pivot Table
pd.pivot_table(df, values='BMI', index='Age', columns='Outcome')
```

```
# Export cleaned dataset for Power BI
df.to_excel('cleaned_healthcare_data.xlsx', index=False)
print("File saved as cleaned_healthcare_data.xlsx")
```
